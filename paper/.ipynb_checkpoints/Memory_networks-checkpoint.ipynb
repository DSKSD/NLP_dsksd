# 2. Memory Network의 주요 Components 

* memory M(an array of objects indexed by $m_i$
* I: (input feature map) – converts the incoming input to the internal feature representation.
* G: (generalization) – updates old memories given the new input. We call this generalization as there is an opportunity for the network to compress and generalize its memories at this stage for some intended future use.
* O: (output feature map) – produces a new output (in the feature representation space), given the new input and the current memory state.
* R: (response) – converts the output into the response format desired. For example, a textual response or an action.


### $I$ Component 

text를 sparse 혹은 dense feature vector로 바꾼다. 이 논문에서는 단순히 BoW로 표현

```python
Image("image/mn2.png")
```

### $G$ Component 

$G$의 가장 간단한 모양은 $I(x)$를 memory의 "slot"에 저장하는 것이다. <br>
$$m_{H(x)} = I(x)$$

slot의 index는 slot을 선택하는 함수인 $H(x)$에 의해 결정된다. 즉 $G$는 단순히 빈칸에 $I(x)$를 집어넣거나 어떤 근거를 가지고 Memory를 update하는 역할을 한다.(이를 위해 $H(x)$는 각 slot에 score를 매길 수도 있다.)

### $O$ and $R$ Components 

$O$는 적절한 답변을 하기 위해 관련 있는 memory가 무엇인지 계산하고 이를 memory로부터 읽어들이는 역할을 한다. $R$은 주어진 $O$로부터 최종 답변을 생성해낸다.(이러한 Task를 위해 $R$은 RNN이 될 수도 있을 것이다.)

### Model 동작 흐름 

1. $x$(an input character, word or sentence)를 internal feature representation $I(x)$로 변환한다.
2. 주어진 new input으로 $m_i$를 업데이트한다. $m_i = G(m_i, I(x),m), $
3. 주어진 new input과 memory로부터 output features $o$를 계산한다. $o = O(I(x),m)$
4. output features $o$를 최종 reponse로 디코딩한다. $ r=R(o)$

# 3. MemNN : Memory neural networks 

## 3.1 Basic model 

#### $I, G$

이 간단한 모델에서는 인풋이 sentence로 들어온다. 그리고 메모리는 새로 들어온 인풋을 그 다음 인덱스에 저장하고, 오래된 메모리는 업데이트 되지 않는다. (더 복잡한 모델은 뒤에서 다루기로 하자, $O,R$의 동작원리를 위해 $I,G$의 모듈을 단순화해서 본다)<br><br>
$$ slot N: m_N = x, N=N+1$$

#### $O$

$O$ 모듈은 주어진 인풋 $x$를 가장 잘 뒷받침하는 k개의 output features를 찾는다. (여기선 k를 2로 뒀음) $s_O$는 sentence $x$와 $m_i$가 얼마나 매치되는지 스코어링 하는 함수일 때, $o_1$는 이 score가 가장 커지는 memory의 인덱스이다. $o_2$는 $x$와 $o_1$과의 매칭을 본다.<br><br>
$$ o_1 = O_1(x,m) = argmax_{i=1,...,N}s_O(x,m_i)$$ <br> 
$$ o_2 = O_2(x,m) = argmax_{i=1,...,N}s_O([x,m_{o_1}],m_i)$$ <br>
이렇게 하여 최종 output $o$는 $[x,m_{o_1},m_{o_2}]$가 된다.

#### $R$

최종적으로 $R$은 textual response $r$을 생성한다. 가장 간단한 방법은 그냥 $m_{o_k}$를 답변으로 돌려주는 것이다. (단순히 검색 결과를 돌려주는 셈) 문장 생성을 위해 RNN을 이용하는 방법도 있지만, 여기선 단순히 가장 적절한 single word를 택하는 방식으로 한다.(이 역시 scoring을 통한 ranking) $W$는 word dictionary이고 $s_R$은 match scoring function <br>
<br>
$$r= argmax_{w\in W}s_R([x,m_{o_1},m_{o_2}],w)$$

```python
Image("image/mn1.png")
```

위를 예를 다시보면, $x$ = "Where is the milk now?"에 답하기 위해, $O$ 모듈은 먼저 모든 메모리에 점수를 매긴다. 가장 관련 있는 fact인 $m_{o_1}$ = "Joe left the milk"를 찾고, 이에 기반하여 그 다음 관련 있는 $m_{o_2}$ = "Joe travelled to the office"를 찾아낸다. 마지막으로 $R$ 모듈은 $[x,m_{o_1},m_{o_2}]$가 주어졌을 때, $r$ = "office"를 찾아낸다.

### Score function

$s_O$와 $s_R$의 포맷은 아래와 같다. <br><br>
$$s(x,y) = g_x(x)^TU^TUg_y(y)$$

$U$는 $n$x$D$의 matrix이다.(D는 embedding dimension, n은 number of features) $g$의 역할은 original text를 $D$-dimensional feature space로 매핑하는 것이다. $D=3|W|$로 정했다. (BoW인데 모든 dictionary는 3가지의 표현이 있는 셈. 하나는 $g_y$를 위해, 두가지는 $g_x$를 위해) $s_O$와 $s_R$은 각각  $U_O$와 $U_R$를 학습한다.

## 3.2 Word Sequences as Input 

만약 인풋이 문장 단위가 아니라 단어의 stream(RNN처럼)이며, 서술과 질문이 구분이 안돼있다면, 또 다른 접근이 필요하다. <br>
이를 위해 "segmentation" function을 추가한다. 지금까지의 시퀀스를 인풋으로 넣어서 중단점(문장의 끝)을 찾아낸다. 만약 중단점을 찾아내서 하나의 segement로 판별된다면 그제서야 memory에 write한다. 

$$seg(c) = {W_{seg}}^TU_Sg_{seg}(c)$$

The segmenter is modeled similarly to our other components, as an embedding model of the form, where $W_{seg}$ is a vector(effectively the parameters of a linear classifier in embedding space), and $c$ is the sequence of input words represened as bag of words using a separate dictionary. If $seg(c) > \gamma$, then this sequence is recognised as a segment.

## 3.3 Efficient Memory via Hashing 

만약 메모리에 저장될 내용이 아주 커지게 되면 앞선 방식으로 score를 계산하는 것이 엄청나게 비싸질 것이다. 이를 위해 <strong>hashing tricks</strong>을 고안했다. 여기에는 2가지 방법이 있는데 (1) via hasing words; via clustering word embeddings. 하지만 첫번째 방식은 문제가 있어 두 번째 방법인 word embedding을 이용한 clustering을 했다. 즉 word vectors를 K-means해서 K개의 buckets에 담았다. (나중에 score 계산할 때 K개의 slot만 고려하면 되기 때문에 계산량이 줄어든다.)

## 3.4 Modeling Write Time 

이제는 "언제" memory slot에 쓸 것인지도 고려하여 모델을 확장한다. 이는 "What is the capital of France?" 와 같은 고정된 사실에 답하는데는 별로 중요하지 않지만, 위의 예시와 같이 story에 관한 질문에 답할 때는 중요하다. <strong>추가 설명 필요</strong>

## 3.5 Modeling Previously Unseen Words 

학습 때 한번도 보지 않은 단어 (Out of Vocaburary)에 대처하기 위해 language model의 아이디어를 빌려온다. <strong>그 단어의 주변 단어를 보고 그 단어를 판단한다.</strong> 이를 위해 feature representation D를 3|W|에서 5|W|로 늘린다. (각 단어 당 오른쪽 왼쪽 주변 단어 역시 feature로 사용하기 위해 2칸 늘린 셈) 학습 시 dropout의 idea를 차용하여 d% 확률로 그 단어를 처음 봤다 가정하고 그 단어가 아닌 주변 context를 사용한다. 

## 3.6 Exact Matches and Unseen words 

임베딩 모델은 n의 낮은 차원 때문에 정확히 단어를 매칭시키진 못한다. 하나의 솔루션은 "bag of words" 매칭 스코어를 추가하는 것이다. <br><br>
$$g_x(x)^TU^TUg_y(y)+\lambda g_x(x)^Tg_y(y)$$ <br>
matching feature는 그 단어가 $x$와 $y$에 둘다 있는지 보는 것이다. 즉, $x$와 $y$가 매칭되는 단어가 있다면 1을 얻을 것이다.
