### Linear chain CRF feature function $f_k(y_t,y_{t-1},x_t)$ 

feature function은 정의하기 나름이지만 가장 심플한 선택은 HMM-like feature set이다.

#### label-label features

$f_{ij}^{LL}(y_t,y_{t-1},x_t) = 1_{\{y_t=i\}}1_{\{y_{t-1}=j\}} $ // 모든 i,j 는 y에 포함된다

이 문제를 풀기 위해, 9개의 다른 labels이 존재하고 그렇기 때문에 81개의 label-label feautres가 있다.

#### label-word featues

$f_{iv}^{LW}(y_t,y_{t-1},x_t) = 1_{\{y_t=i\}}1_{\{x_t=v\}}$ // i는 Y에 포함, v는 V(Vocab)에 포함

#### full set of features $F$ :는 위 2 features의 합집합

```python
import numpy as np
import tensorflow as tf
```

```python
# Data settings.
num_examples = 10
num_words = 20
num_features = 100
num_tags = 5

# Random features.
x = np.random.rand(num_examples, num_words, num_features).astype(np.float32)

# Random tag indices representing the gold sequence.
y = np.random.randint(num_tags, size=[num_examples, num_words]).astype(np.int32)

# All sequences in this example have the same length, but they can be variable in a real model.
sequence_lengths = np.full(num_examples, num_words - 1, dtype=np.int32)
```

```python
# Add the data to the TensorFlow graph.
x_t = tf.constant(x)
y_t = tf.constant(y)
sequence_lengths_t = tf.constant(sequence_lengths)
```

```python
# Compute unary scores from a linear layer.
weights = tf.get_variable("weights", [num_features, num_tags])
matricized_x_t = tf.reshape(x_t, [-1, num_features]) # 10,20,100 -> 200x100
matricized_unary_scores = tf.matmul(matricized_x_t, weights) # 200x100 x 100 x5 => 200 x 5
unary_scores = tf.reshape(matricized_unary_scores,
                              [num_examples, num_words, num_tags]) #  200x5 -> 10,20,5
```

```python
# Compute the log-likelihood of the gold sequences and keep the transition
# params for inference at test time.
log_likelihood, transition_params = tf.contrib.crf.crf_log_likelihood(
        unary_scores, y_t, sequence_lengths_t)
```

```python
# Add a training op to tune the parameters.
loss = tf.reduce_mean(-log_likelihood)
train_op = tf.train.GradientDescentOptimizer(0.01).minimize(loss)
```

```python
# Train and evaluate the model.
with tf.Graph().as_default():
  with tf.Session() as session:

    # Train for a fixed number of iterations.
    session.run(tf.global_variables_initializer())
    for i in range(1000):
      tf_unary_scores, tf_transition_params, _ = session.run(
          [unary_scores, transition_params, train_op])
      if i % 100 == 0:
        correct_labels = 0
        total_labels = 0
        for tf_unary_scores_, y_, sequence_length_ in zip(tf_unary_scores, y,
                                                          sequence_lengths):
          # Remove padding from the scores and tag sequence.
          tf_unary_scores_ = tf_unary_scores_[:sequence_length_]
          y_ = y_[:sequence_length_]

          # Compute the highest scoring sequence.
          viterbi_sequence, _ = tf.contrib.crf.viterbi_decode(
              tf_unary_scores_, tf_transition_params)

          # Evaluate word-level accuracy.
          correct_labels += np.sum(np.equal(viterbi_sequence, y_))
          total_labels += sequence_length_
        accuracy = 100.0 * correct_labels / float(total_labels)
        print("Accuracy: %.2f%%" % accuracy)
```

```python
sents = []
tags = []

for i in range(len(data)):
    temp_w = []
    temp_t = []
    for word in data[i]:
        try:
            int(word[0])
            temp_w.append('<NUM>')
        except:
            temp_w.append(word[0])
        temp_t.append(word[2])

    sents.append(temp_w)
    tags.append(temp_t)
```

```python
num_examples = 649
num_words = 50
num_features = 50 # embedding
num_tags = 14
sequence_lengths = []
vocab_size = 1515
```

```python
exceptid = 0
for i in range(len(sents)):
    if len(sents[i]) == 0: 
        exceptid = i
        continue
    sequence_lengths.append(len(sents[i]))
    if i % 100 == 0:
        print(i)
    
    while len(sents[i]) != 50:
        sents[i].append('<PAD>')
        tags[i].append('<DUM>')
    i+=1
```

```python
Vocab = list(set(flatten(sents)))
tagset = list(set(flatten(tags)))
```

```python
sequence_lengths = np.array(sequence_lengths, dtype=np.int32)
sequence_lengths.shape
```

```python
def sents_to_index(sents):
    result=[]
    for sent in sents:
        temp = []
        for word in sent:
            index = Vocab.index(word)
            temp.append(index)
        result.append(temp)
    
    return result
```

```python
def tags_to_index(tags):
    result = []
    for tag in tags:
        temp = []
        for word in tag:
            index = tagset.index(word)
            temp.append(index)
        result.append(temp)
    
    return result
```

```python
X_data = sents_to_index(sents)
y_data = tags_to_index(tags)
X_data = flatten(X_data)
X_data = np.array(X_data)
y_data = np.array(y_data)
```

```python
X_data.shape
```

```python
y_data.shape
```

```python
input_words = tf.placeholder(tf.int32, shape=[num_examples*num_words], name='input_words')
y = tf.placeholder(tf.int32, shape=[num_examples,num_words], name='y')
sequence_lengths_t = tf.constant(sequence_lengths)
```

```python
embed_matrix = tf.Variable(tf.random_uniform([vocab_size, 
                                                                    num_features], -1.0, 1.0), 
                                                                    name='embed_matrix')

matricized_x_t = tf.nn.embedding_lookup(embed_matrix, input_words, name='embed') # nx50
weights = tf.Variable(tf.random_uniform([num_features, num_tags]
                                        , -1.0, 1.0),dtype=tf.float32,name='crf_weights') # 50x14

matricized_unary_scores = tf.matmul(matricized_x_t, weights) # nx50 x 50 x14 => n x 14
unary_scores = tf.reshape(matricized_unary_scores,
                             [num_examples, num_words, num_tags]) #  nx14 -> 650,50,14
```

```python
log_likelihood, transition_params = tf.contrib.crf.crf_log_likelihood(
        unary_scores, y, sequence_lengths_t)
```

```python
# Add a training op to tune the parameters.
loss = tf.reduce_mean(-log_likelihood)
train_op = tf.train.GradientDescentOptimizer(0.01).minimize(loss)
```

```python
# Train and evaluate the model.
with tf.Session() as session:

    # Train for a fixed number of iterations.
    session.run(tf.global_variables_initializer())

    #session.run(loss, feed_dict={input_words : X_data, y : y_data})
    for i in range(1000):
      tf_unary_scores, tf_transition_params, _ = session.run(
          [unary_scores, transition_params, train_op],feed_dict={input_words : X_data, y : y_data})
      if i % 100 == 0:
        print(tf_unary_scores.shape)
        correct_labels = 0
        total_labels = 0
        for tf_unary_scores_, y_, sequence_length_ in zip(tf_unary_scores, y_data,
                                                          sequence_lengths):
          # Remove padding from the scores and tag sequence.
          tf_unary_scores_ = tf_unary_scores_[:sequence_length_]
          y_ = y_[:sequence_length_]
          # Compute the highest scoring sequence.
          viterbi_sequence, _ = tf.contrib.crf.viterbi_decode(
              tf_unary_scores_, tf_transition_params)

          # Evaluate word-level accuracy.
          correct_labels += np.sum(np.equal(viterbi_sequence, y_))
          total_labels += sequence_length_
        accuracy = 100.0 * correct_labels / float(total_labels)
        print("Accuracy: %.2f%%" % accuracy)
```
